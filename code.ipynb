## Loading the model

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Create a pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False,
)
```

```python
import pandas as pd
# Read the .jsonl file
df = pd.read_json('train.jsonl', lines=True)
df_subset = df.iloc[:100]  # consider only first 100 rows
df_subset
```

## One-shot Chain-of-Thought


```python
# # Answering without explicit reasoning

standard_prompt = [
     {"role": "user", "content": "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"},
     {"role": "assistant", "content": "11"},
     {"role": "user", "content": "The cafeteria had 25 apples. If they used 20 to make lunch and bought 6 more, how many apples do they have?"}
 ]

# # Run generative model
outputs = pipe(standard_prompt)
print(outputs[0]["generated_text"])
```

```python
# Create a list to store the outputs
results = []

# Iterate over each row in the dataframe
for question in df_subset["question"]:
    # Construct the prompt using the current question
    standard_prompt = [
        {"role": "user", "content": "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"},
        {"role": "assistant", "content": "11"},
        {"role": "user", "content": f"{question}"},  # Use the current question in the dataframe
    ]

    # Generate the output for the question using the generative model
    outputs = pipe(standard_prompt)

    # Extract the generated response (assuming it's stored in "generated_text")
    response = outputs[0]["generated_text"]

    # Append the question and response to the results list
    results.append({"question": question, "response": response})

# Convert the results into a DataFrame
output_df = pd.DataFrame(results)

# Display the results in tabular format
import IPython.display as display
display.display(output_df)
```

##Self-Consistency with CoT

```python
# Create a list to store the outputs
results = []

# Iterate over each row in the dataframe
for question in df_subset["question"]:
    # Construct the prompt with chain-of-thought reasoning and self-consistency
    standard_prompt = [
        {"role": "user", "content": "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"},
        {"role": "assistant", "content": "Let's break it down step by step. Roger started with 5 tennis balls. He buys 2 cans, and each can contains 3 tennis balls. So, 2 cans × 3 tennis balls per can = 6 tennis balls. Adding 5 tennis balls to 6 tennis balls gives us 11. The total number of tennis balls Roger has now is 11."},
        {"role": "user", "content": f"{question}"},  # Use the current question from the dataframe
        {"role": "assistant", "content": "Let's think through this carefully. First, we need to break down the problem step by step, then check the results to ensure accuracy."}
    ]

    # Generate multiple responses (self-consistency)
    outputs = []
    for _ in range(5):  # Generate 5 different responses to check for consistency
        output = pipe(standard_prompt)
        response = output[0]["generated_text"]
        outputs.append(response)

    # Find the most frequent (or consistent) response
    from collections import Counter
    response_counter = Counter(outputs)
    most_common_response = response_counter.most_common(1)[0][0]

    # Append the question and the most common response to the results list
    results.append({"question": question, "response": most_common_response})

# Convert the results into a DataFrame
output_df = pd.DataFrame(results)

# Display the results in tabular format
import IPython.display as display
display.display(output_df)
```
